# Part 2 – Statistical Analysis


## Pattern 1 – Relationship between GDP and Happiness Score

I want to investigate if whether countries with higher GDP per capita tend to have higher happiness scores.

```python
gdp_sorted = df.sort_values(by='GDP per capita', ascending=False)
```

```python
print(gdp_sorted[['Country or region', 'GDP per capita', 'Score']].head())
```

```python
print(df[['GDP per capita', 'Score']].corr())
```

**Write here your statistical discussion of Pattern 1.**

The results show a clear positive relationship between GDP per capita and happiness score. Countries like Qatar, Luxembourg, and Singapore have some of the highest GDP values and also relatively high happiness scores. The correlation value of **0.79** indicates a strong positive relationship, meaning that as GDP per capita increases, happiness scores tend to increase as well.

This is interesting because it suggests that economic prosperity plays an important role in overall happiness. However, since the correlation is not perfect (not equal to 1), it also implies that wealth alone does not fully determine happiness — other factors such as social support and quality of life also contribute significantly.


## Pattern 2 – Social support vs Happiness

I want to investigate if countries with stronger social support systems have higher happiness scores.

```python
support_sorted = df.sort_values(by='Social support', ascending=False)

print(support_sorted[['Country or region', 'Social support', 'Score']].head())
```

```python
print(df[['Social support', 'Score']].corr())
```

Statistical discussion of pattern 2 :

The results show a strong positive relationship between social support and happiness score. Countries such as Iceland, Finland, and Norway have very high levels of social support and also rank among the happiest countries. The correlation value of **0.78** indicates a strong link, meaning that countries where people feel more supported by family, friends, or society tend to report higher happiness.

This is interesting because it highlights that happiness is not just about wealth or economic factors. Strong social connections and support systems play a major role in people’s well-being. However, since the correlation is not perfect, it suggests that while social support is important, other factors like income, health, and freedom also contribute to overall happiness.


# Part 3 – Visualizations

## Plot 1

Plot 1 shows how GDP affects Happiness Score...

```python
df_sorted_gdp = df.sort_values(by='GDP per capita')

plt.figure()
plt.plot(df_sorted_gdp['GDP per capita'], df_sorted_gdp['Score'])
plt.title("GDP per Capita vs Happiness Score")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.show()
```


**Write here a short comment explaining what Plot 1 shows and why it is interesting.**


This plot shows how happiness changes as GDP per capita increases. There is a clear upward trend, meaning countries with higher GDP generally have higher happiness scores. This supports the earlier correlation result (0.79), confirming a strong positive relationship. However, the line is not perfectly straight, which suggests that while wealth plays a major role, it is not the only factor influencing happiness.

## Plot 2

Plot 2 shows how Social Support affects Happiness Score. ....

```python
import seaborn as sns

sns.pairplot(df[['Social support', 'Score']])
plt.show()
```


**Write here a short comment explaining what Plot 2 shows and why it is interesting.**


This plot shows the relationship between social support and happiness. The points form a clear upward pattern, indicating that countries with stronger social support systems tend to have higher happiness scores. This aligns with the correlation value (0.78), showing a strong positive relationship. It is interesting because it highlights that emotional and social well-being are just as important as economic factors in determining happiness

## Plot 3

Happiness Score by GDP Category

```python
df['GDP_category'] = pd.qcut(df['GDP per capita'], q=3, labels=['Low', 'Medium', 'High'])

plt.figure()
df.boxplot(column='Score', by='GDP_category')
plt.title("Happiness Score by GDP Category")
plt.suptitle("")
plt.xlabel("GDP Category")
plt.ylabel("Happiness Score")
plt.show()
```


**Write here a short comment explaining what Plot 3 shows and why it is interesting.**


This plot groups countries into low, medium, and high GDP categories and compares their happiness scores. Countries in the high GDP group clearly have higher median happiness scores, while low GDP countries tend to have lower scores. The spread also shows that high GDP countries are more consistently happy. This is interesting because it reinforces the idea that economic strength contributes to happiness, but also shows some variation, meaning not all wealthy countries are equally happy.
